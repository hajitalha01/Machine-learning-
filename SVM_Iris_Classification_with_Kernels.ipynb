**Libraries Import**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Dataset Load**

```python
df = pd.read_excel("Iris.xlsx")
df.head()
```

**Dataset Info**

```python
df.info()
```

**Check Missing Values**

```python
df.isnull().sum()
```

**Id Column Remove**

```python
df = df.drop("Id", axis=1)
df.head()
```

**Features aur Target Separate**

```python
X = df.drop("Species", axis=1)
y = df["Species"]
```

**Task 1 — SVM with 2 Features**

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# 2 features select
X2 = df[['SepalLengthCm', 'SepalWidthCm']]
y = df['Species']

# split
from sklearn.model_selection import train_test_split
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y, test_size=0.2, random_state=42)

# model
svm_2 = SVC(kernel='linear')

svm_2.fit(X_train2, y_train2)

# prediction
pred2 = svm_2.predict(X_test2)

# accuracy
acc2 = accuracy_score(y_test2, pred2)

print("Accuracy with 2 features:", acc2)
```

**Task 2 — SVM with 4 Features**

```python
X4 = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = df['Species']

X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y, test_size=0.2, random_state=42)

svm_4 = SVC(kernel='linear')

svm_4.fit(X_train4, y_train4)

pred4 = svm_4.predict(X_test4)

acc4 = accuracy_score(y_test4, pred4)

print("Accuracy with 4 features:", acc4)
```

**Task 3 — Different Kernels**

```python
svm_linear = SVC(kernel='linear')

svm_linear.fit(X_train4, y_train4)

pred_linear = svm_linear.predict(X_test4)

acc_linear = accuracy_score(y_test4, pred_linear)

print("Linear Kernel Accuracy:", acc_linear)
```

**Polynomial Kernel**

```python
svm_poly = SVC(kernel='poly')

svm_poly.fit(X_train4, y_train4)

pred_poly = svm_poly.predict(X_test4)

acc_poly = accuracy_score(y_test4, pred_poly)

print("Polynomial Kernel Accuracy:", acc_poly)
```

**RBF Kernel**

```python
svm_rbf = SVC(kernel='rbf')

svm_rbf.fit(X_train4, y_train4)

pred_rbf = svm_rbf.predict(X_test4)

acc_rbf = accuracy_score(y_test4, pred_rbf)

print("RBF Kernel Accuracy:", acc_rbf)
```

**Task 4 — Compare Results**

```python
print("Comparison of Models")

print("SVM with 2 Features:", acc2)
print("SVM with 4 Features:", acc4)

print("Linear Kernel:", acc_linear)
print("Polynomial Kernel:", acc_poly)
print("RBF Kernel:", acc_rbf)
```
